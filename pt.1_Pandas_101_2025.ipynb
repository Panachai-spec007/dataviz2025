## HW 5.1

````
# prompt:  row  column description

folk_df = folk_df.dropna(subset=['description'])
````

````
# prompt:  html tags  html codes  text  column description  beautifulsoup

from bs4 import BeautifulSoup

def remove_html_tags(text):
  soup = BeautifulSoup(text, "html.parser")
  cleaned_text = soup.get_text()
  return cleaned_text

folk_df['description'] = folk_df['description'].apply(remove_html_tags)
````

````
folk_df.head()
````

````
folk_df['description'][0]
````

````
# prompt:  \xa0  \n 

def remove_special_chars(text):
  cleaned_text = text.replace('\xa0', '')
  cleaned_text = cleaned_text.replace('\n', '')
  return cleaned_text

folk_df['description'] = folk_df['description'].apply(remove_special_chars)
````

````
folk_df['description'][0]
````

````
# prompt: description  row  print name  

longest_description = ''
longest_description_name = ''
for index, row in folk_df.iterrows():
  if len(row['description']) > len(longest_description):
    longest_description = row['description']
    longest_description_name = row['title']

print(f'The longest description belongs to "{longest_description_name}" and its length is {len(longest_description)}')
````

````
# prompt: description  row  print name  

shortest_description = None
shortest_description_name = None
for index, row in folk_df.iterrows():
  if shortest_description is None or len(row['description']) < len(shortest_description):
    shortest_description = row['description']
    shortest_description_name = row['title']

print(f'The shortest description belongs to "{shortest_description_name}" and its length is {len(shortest_description)}')
````

````
# prompt:     

import pandas
import pandas as pd
import os
from google.colab import drive
from bs4 import BeautifulSoup

# ... (preceding code)

count_life = 0
count_local = 0
for index, row in this_data.iterrows():
  if ('' in row['title'] or '' in row['title']) and row['type'] == '':
    count_life += 1
  elif ('' in row['title'] or '' in row['title']) and row['type'] == '':
    count_local += 1

if count_life > count_local:
  print(f"There are more stories with '' or '' in the title in  ({count_life} stories) than in  ({count_local} stories).")
elif count_local > count_life:
  print(f"There are more stories with '' or '' in the title in  ({count_local} stories) than in  ({count_life} stories).")
else:
  print(f"There are an equal number of stories with '' or '' in the title in both  and  ({count_life} stories each).")
````

````
# prompt:        ( code )

#  dictionary  ''  '' 
creator_counts = {}
for index, row in this_data.iterrows():
  if '' in row['title'] or '' in row['title']:
    creator = row['creator']
    if creator not in creator_counts:
      creator_counts[creator] = 0
    creator_counts[creator] += 1

# 
max_count = 0
most_prolific_creator = None
for creator, count in creator_counts.items():
  if count > max_count:
    max_count = count
    most_prolific_creator = creator

print(f"The creator who wrote the most stories with '' or '' in the title is {most_prolific_creator} with {max_count} stories.")
````
